## LINEAR REGRESSION

```python
# moving to the linear regression
from sklearn.linear_model import LinearRegression
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train,y_train)
```

```python
for idx,col_name  in enumerate(x_train.columns):
    print("The  coefficient for {} is {}".format(col_name,regression_model.coef_[0][idx]))

# in here i'm using this concept because i have a n number features (Multi linear regression)
# using for loop and enumerate function finding the features values
# this is  my CO-efficient
```

```python
intercept = regression_model.intercept_[0]
print("The  intercept  for our model is {}".format(intercept))
# here as same i'm checking with the intercept value like my  traget
# this is my intercept value
```

```python
score = regression_model.score(x_train,y_train)
print("Score R2 :",score)
# here my score r2 value is 
```

```python
# finding the mean square error
mse = np.mean((regression_model.predict(x_test)-y_test)**2)
print("Mean Square Error :",mse)
```

```python
# finding a root of mean square  error
rmse = np.sqrt(mse)
print("Root Of Mean Square Error :",rmse)
```

```python
adj_r2 = 1-(1-score)*(len(y_test) - 1) / (len(y_test) - x_test.shape[1] - 1)
print("Adjusted R square :",adj_r2)
```
